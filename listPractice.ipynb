```python
#Making a list

name = ["tom", "bob", "alex", "jim", "Mia", "eva", "jason", "harry", "jack"]
print(name)
print()

#accesing the element

print("first element: ", name[0])
print("last element: ", name[-1])
print()

#modifing indivisual items
print("After modification")
print("--------------------------------------------------------------------------")
name[0] = 'jade'
name[1] = 'ronald'
print(name)
print()

#adding elements
print("after adding elements")
print("--------------------------------------------------------------------------")
name.append('tom')
name.append('bob')
print(name)
print()

#inserting elements at perticular position
print("inserting elements at position 2,5")
print("--------------------------------------------------------------------------")
name.insert(1, 'mark')
name.insert(4, 'Joe')
print(name)
print()

#deleting elements by it's position
print("after deleting at position 2,5")
print("--------------------------------------------------------------------------")
del name[1]
del name[4]
print(name)
print()

#removing an item by value
print("After removing Mia from name")
print("--------------------------------------------------------------------------")
name.remove('Mia')
print(name)
print()

#popping elements
print("popping last item")
print("--------------------------------------------------------------------------")
name.pop()
print(name)
print()
print("popping first item")
print("--------------------------------------------------------------------------")
name.pop(0)
print(name)
print()

#Length of a list
print("List Length")
print("--------------------------------------------------------------------------")
name_len = len(name)
print("length of a list name is: ", name_len )
print()

#Sorting List
print("sorting list")
print("--------------------------------------------------------------------------")
name.sort()
print(name)
print()
#Sorting List in reverse order
print("sorting list in reverse  order")
print("--------------------------------------------------------------------------")
name.sort(reverse=True)
print(name)
print()

#Looping through the list
print("Looping through list")
print("--------------------------------------------------------------------------")
for i in name:
    print(i)
print()
#Printing msg for each item
print("Msg for each otem")
print("--------------------------------------------------------------------------")
for i in name:
    print("welcome " + i + "!")
```
